```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("summer.csv")
```

```python
df.head()
```

### 1. In how many cities Summer Olympics is held so far?

```python
city=df['City'].unique()
print(len(city))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df_gold=df[df['Medal']=='Gold']
data=[]

for sport in df_gold['Sport'].unique():
    data.append([sport, len(df_gold[df_gold['Sport']==sport])])

print(pd.DataFrame(data,columns=['Sport','Gold Medal']).sort_values(by='Gold Medal',ascending=False).head())
pd.DataFrame(data,columns=['Sport','Gold Medal']).sort_values(by='Gold Medal',ascending=False).head().plot(x='Sport',y='Gold Medal', kind='bar',figsize=(10,5),title='Sports having most number of Gold Medals',ylabel='No. of Gold Medals',color='#FFAF3D')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df_medal=df[df['Medal'].isin(['Gold', 'Silver', 'Bronze'])]

data=[]

for sport in df_medal['Sport'].unique():
    data.append([sport, len(df_medal[df_medal['Sport']==sport])])

print(pd.DataFrame(data,columns=['Sport','Medal']).sort_values(by='Medal',ascending=False).head())
pd.DataFrame(data,columns=['Sport','Medal']).sort_values(by='Medal',ascending=False).head().plot(x='Sport',y='Medal', kind='bar',figsize=(10,5),title='Sports having most number of Medals',ylabel='No. of Medals',color='#6AB9BB')
```

### 4. Which player has won most number of medals? (Top 5)

```python
df_medal=df[df['Medal'].isin(['Gold', 'Silver', 'Bronze'])]

data=[]

for athlete in df_medal['Athlete'].unique():
    data.append([athlete, len(df_medal[df_medal['Athlete']==athlete])])

print(pd.DataFrame(data,columns=['Player','Medal']).sort_values(by='Medal',ascending=False).head())
pd.DataFrame(data,columns=['Player','Medal']).sort_values(by='Medal',ascending=False).head().plot(x='Player',y='Medal', kind='bar',figsize=(10,5),title='Players winning most number of Medals',ylabel='No. of Medals',color='#A56ABB')
```

```python
# using groupby

df.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot(kind='bar',figsize=(10,5),title='Players winning most number of Medals',ylabel='No. of Medals',color='#6A74BB')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df_gold=df[df['Medal'].isin(['Gold'])]

data=[]

for athlete in df_gold['Athlete'].unique():
    data.append([athlete, len(df_gold[df_gold['Athlete']==athlete])])

print(pd.DataFrame(data,columns=['Player','Gold Medal']).sort_values(by='Gold Medal',ascending=False).head())
pd.DataFrame(data,columns=['Player','Gold Medal']).sort_values(by='Gold Medal',ascending=False).head().plot(x='Player',y='Gold Medal', kind='bar',figsize=(10,5),title='Players winning most number of Gold Medals',ylabel='No. of Gold Medals',color='#6ABB77')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df_gold=df[(df['Medal']=='Gold') & (df['Country']=='IND')]
data=[]

for year in df_gold['Year'].unique():
    data.append(year)

data.sort()
print(data[0])
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]

for event in df['Event'].unique():
    data.append([event, len(df[df['Event']==event])])

print(pd.DataFrame(data,columns=['Event','Players']).sort_values(by='Players',ascending=False).head())
pd.DataFrame(data,columns=['Event','Players']).sort_values(by='Players',ascending=False).head().plot(x='Event',y='Players', kind='bar',figsize=(10,5),title='Event most popular in terms on number of Players',ylabel='No. of Players',color='#BB6A91')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df_gold=df[(df['Medal']=='Gold') & (df['Gender']=='Women')]

data=[]

for sport in df_gold['Sport'].unique():
    data.append([sport, len(df_gold[df_gold['Sport']==sport])])

print(pd.DataFrame(data,columns=['Sport','Female Gold Medalist']).sort_values(by='Female Gold Medalist',ascending=False).head())
pd.DataFrame(data,columns=['Sport','Female Gold Medalist']).sort_values(by='Female Gold Medalist',ascending=False).head().plot(x='Sport',y='Female Gold Medalist', kind='bar',figsize=(10,5),title='Sports having most Female Gold Medalists',ylabel='No. of Female Gold Medalists',color='#BB6C6A')
```
